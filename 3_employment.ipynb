```python
## calculate change in employment reported, year over year

import csv, operator

fn = "Current_Employment_Statistics__Beginning_1990.csv"
f = open(fn, 'r')

JurisCount = 0
reader = csv.reader(f)

## create dictionaries for year 1 and year 2

EmpHist = {}
EmpHist2 = {}
# n is counting every row in reader
n = 0
# n2 is counting rows which are y = 2020 and d not null
n2 = 0

## designate the first year, yr 2 is calculated
yr1 = '2020'
yr2 = str(int(yr1) + 1)

for row in reader :
    j = row[1]
    y = row[4]
    m = row[5]
    ## employment
    e = row[6]
    ## type
    t = row[3]
    
    n += 1
    
    ## test
    #print(n)
    #if n > 20 :
    #    break
    
    #print(n, j, m, y, t, e)
    
    ## create a running sum (data is by month, this totals for the year) for the first year - only "total nonfarm" type employment 
    ##    this will be averaged
    ## to compare to sales tax data, translate the years - January and February of year 1 are part of the sales tax year, as are the
    ##   months March through December of the previous year
    if y == yr1 and m <= '2' and t == 'Total Nonfarm' and e != '' :
        ## test
        #print(n, j, m, y, t, e)
        if j in EmpHist.keys() :
            employ = int(EmpHist[j])
            add = int(e)
            EmpHist[j] = employ + add
        else:
            EmpHist[j] = int(e)
    if y == str(int(yr1)-1) and m > '2' and t == 'Total Nonfarm' and e != '' :
        ## test
        #print(n, j, m, y, t, e)
        if j in EmpHist.keys() :
            employ = int(EmpHist[j])
            add = int(e)
            EmpHist[j] = employ + add
        else:
            EmpHist[j] = int(e)
    
    ## create a running sum for the second year - only "total nonfarm" type employment, this will be averaged
    if y == yr2 and m <= '2' and t == 'Total Nonfarm' and e != '' :
        ## test
        #print(n, j, m, y, t, e)
        if j in EmpHist2.keys() :
            employ = int(EmpHist2[j])
            add = int(e)
            EmpHist2[j] = employ + add
        else:
            EmpHist2[j] = int(e)
    if y == str(int(yr2)-1) and m > '2' and t == 'Total Nonfarm' and e != '' :
        ## test
        #print(n, j, m, y, t, e)
        if j in EmpHist2.keys() :
            employ = int(EmpHist2[j])
            add = int(e)
            EmpHist2[j] = employ + add
        else:
            EmpHist2[j] = int(e)
            
## average the total employment numbers for the year to a monthly rate
##  each month is not additional jobs, so to compare to sales tax numbers for the entire year
##  decided to average the jobs for the same time period
for key in EmpHist.keys() :
    #print(key, EmpHist[key])
    TotalEmp = EmpHist[key]
    EmpHist[key] = TotalEmp/12
    #print(TotalEmp, key, EmpHist[key])
for key1 in EmpHist2.keys() :
    #print(key1, EmpHist2[key1])
    TotalEmp2 = EmpHist2[key1]
    EmpHist2[key1] = TotalEmp2/12
    #print(TotalEmp2, key1, EmpHist2[key1])
        
#print("Year 1:", yr1, EmpHist)
#print("Year 2:", yr2, EmpHist2)


print("Average employment reported by year, by jurisdiction, Total Nonfarm type employment")

## create lists to store any jurisdictions reported in 1 year but not the other
Yr1_only = []
Yr2_only = []
for key1 in EmpHist.keys():
    if key1 not in EmpHist2.keys():
        Yr1_only.append(key1)
for key2 in EmpHist2.keys():
    if key2 not in EmpHist2.keys() :
        Yr2_only.append(key2)
print("Jurisdictions in Yr1 only:", Yr1_only)
print("Jurisdictions in Yr2 only:", Yr2_only)

## create a dictionary for the employment change, to compare with the sales tax data
Emp_change = {}

## print the average employment by year, change between years, and percentage change
print(f"{'Jurisdiction':40}{yr1:>20}{yr2:>20}{'Change':>20}{'Percent change':>20}")
for key, value in EmpHist.items() :
    change = EmpHist2[key] - EmpHist[key]
    perc_change = change/EmpHist[key]
    print(f"{key:40}{value:>20,.0f}{EmpHist2[key]:>20,.0f}{change:>20,.0f}{perc_change:>20.2%}")
    Emp_change[key] = perc_change
#print(Emp_change)
```

```python
## initial open the st_change file created from the sales tax reported dataset

f_st = open("st_change.csv","r")
s_st = f_st.read()
f_st.close()
lines = s_st.split("\n")
```

```python
## sales tax data, calculated by the same geographical areas as the DOL data
## important to check the results to the results from the sales tax dataset analysis - this can be done by
##    looking at some of the "minor counties", such as Cayuga, Fulton, and Montgomery

import csv, operator

f_st_change = "st_change.csv"
f_st_ch = open(f_st_change, 'r')

f_dol = "DOL_geog_areas.csv"
f_areas = open(f_dol, 'r')

JurisCount = 0
reader_stc = csv.reader(f_st_ch)
reader_dol = csv.reader(f_areas)
## create a dictionary for storing the data
ST_changeHist = {}

#region = ''

n_stc = 0

for row_c in reader_stc:
    reader_dol = csv.reader(open(f_dol,'r'))
    if n_stc > 0 :
        y1 = float(row_c[1])
        y2 = float(row_c[2])
        change = y2 - y1
        base = y1
        area = row_c[0]
        #print(n_stc, area, y1, y2, change, base)
        
        ## find the applicable DOL geographical area for the sales tax jurisdiction
        for row_dol in reader_dol :
            if area == row_dol[1].upper() :
                region = row_dol[3]
                break
                
        ## calculate the change and percentage change, from year 1 to year 2, and store in dictionary
        if region in ST_changeHist.keys () :
            adj_change = int(ST_changeHist[region]['change'])
            add = int(change)
            ST_changeHist[region]['change'] = adj_change + add

            base_change = int(ST_changeHist[region]['base'])
            add_base = int(base)
            ST_changeHist[region]['base'] = base_change + add_base
        else :
            ST_changeHist[region] = {}
            ST_changeHist[region]['change'] = change
            ST_changeHist[region]['base'] = base
        #print(n_stc, area, region, change, base)
    n_stc += 1

#print(ST_changeHist)

## create a dictionary for storing the calculations for comparison to the DOL data
p_change_hist = {}

## print the results and store them in the dictionary p_change_hist
print("Change in sales tax from year 1, by DOL geographical area")
print(f"{'Area':50}{'Change':>20}{'Year 1':>20}{'Percent change':>20}")
for key in ST_changeHist :
    p_change = ST_changeHist[key]['change']/ST_changeHist[key]['base']
    p_change_hist[key] = p_change
    ch = ST_changeHist[key]['change']
    bs = ST_changeHist[key]['base']
    print(f"{key:50}{ch:>20,.2f}{bs:>20,.2f}{p_change:>20,.2%}")
    
#print(p_change_hist)
    
```

```python
## compare the sales tax change to the employment data change

## create lists for any jurisdictions which were in one dictionary, but not the other
Emp_only = []
ST_only = []
## compare dictionaries and store any unmatched
for keyE in Emp_change.keys():
    if keyE not in p_change_hist.keys():
        Emp_only.append(keyE)
for keyST in p_change_hist.keys():
    if keyST not in Emp_change.keys() :
        ST_only.append(keyST)
        
print("Change in employment numbers (DOL) to change in sales tax, year over year")
print("Regions in Employment data only:", Emp_only)
print("Regions in ST data only:", ST_only)

## print the DOL change, ST change, and difference
print(f"{'Region':50}{'Employment change':>20}{'ST change':>20}{'Difference':>20}")
for key, value in Emp_change.items() :
    if key in p_change_hist :
        differ = p_change_hist[key] - Emp_change[key]
        print(f"{key:50}{value:>20,.2%}{p_change_hist[key]:>20,.2%}{differ:>20,.2%}")
```
