```python
##

import csv, operator

fn = "Taxable_Sales_And_Purchases_Quarterly_Data__Beginning_Sales_Tax_Year_2013-2014.csv"
f = open(fn, 'r')

f2 = "New_York_State_Corporate_Tax_Credits_by_Major_Industry_Group__Beginning_Tax_Year_2001.csv"
fc = open(f2, 'r')

s_reader = csv.reader(f)
c_reader = csv.reader(fc)
Sales = {}
Credits = {}
n = 0
n2 = 0

    # NAICS code 4441 is Building Material and Supplies dealers - for example, Home Depot
    # NAICS code 4453 is Beer, Wine, and Liquor Stores
    # NAICS code 7225 is Restaurants and Other Eating Places
    # NAICS code 5511 is Management of Companies and Enterprises - this was the only NAICS code in both taxable sales/purchases and tax credits
    # error handling - code 0000 is where there is a description in the tax credits file which does not have a match in the NAICS code file
# which NAICS code to look for, defined as nc
nc = '5511'

print('NAICS code data: ', nc)

f_naics = open('naics.csv','w')
f_naics.write("")
f_naics.close()

#reader = csv.reader(f)
f_naics = open('naics.csv','a')
#writer = csv.writer(f_naics, delimiter = ',', lineterminator = '\n')

naics_dict = {}

for row1 in s_reader :
    n_d = 0
    ind = row1[5]
    desc = row1[6]
    if n> 0 and desc not in naics_dict.keys() :
        naics_dict[desc] = ind
    n += 1
csv_file = "naics.csv"
try:
    with f_naics as csvfile:
        csv_columns = ['Desc', 'NAICS_code']
        #writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer = csv.writer(csvfile, delimiter = ',', lineterminator = '\n')
        writer.writerow(csv_columns)
        #for data in naics_dict :
        for data in naics_dict.items() :
            writer.writerow(data)
except IOError:
    print("I/O error")

n = 0 
f = open(fn, 'r')
s_reader = csv.reader(f)

for row in s_reader:
    industry = row[5]
    tax_yr = row[1][-4:]
    r = row[7]
    if industry == nc and r!= '' :
        if tax_yr in Sales.keys() :
            rev = int(Sales[tax_yr])
            add = int(r)
            Sales[tax_yr] = rev + add
        else :
            Sales[tax_yr] = int(r)
        #if n % 1000 == 0 :
            #print(n)
        n += 1
print('Taxable Sales and Purchases - dictionary')
print(Sales)        

n = 0
fc = open(f2, 'r')
c_reader = csv.reader(fc)
f_codes = "naics.csv"
f_nc = open(f_codes, 'r')
 
for line in c_reader:
    reader2 = csv.reader(open(f_codes,'r'))
    industry_desc = line[4]
    
    if n > 0 : 
        for code in reader2 :
            list_desc = code[0]
            if list_desc == industry_desc :
                industry = code[1]
                break
            else:
                # error handling - if there is no match, designate as code 0000, description "not found"
                list_desc = 'Not found'
                industry = '0000'
        t_yr = line[0]
        c = line[7]
    
    if industry == nc and c != '' :
    #and n > 1100 and n < 1200:
    #if n > 0: 
        if t_yr in Credits.keys() :
            cred = int(Credits[t_yr])
            add = int(c)
            Credits[t_yr] = cred + add
        else :
            Credits[t_yr] = c
        #print(t_yr, c, industry)
        #if n % 1000 == 0 :
            #print(n)
    n += 1

#print(n)
#print(nc)

print('Tax Credits - dictionary')
print(Credits)
        
Sales_only = []
Credits_only = []
for key1 in Sales.keys():
    if key1 not in Credits.keys():
        Sales_only.append(key1)
for key2 in Credits.keys():
    if key2 not in Sales.keys() :
        Credits_only.append(key2)
print("Years in Taxable Sales and Purchases only:", Sales_only)
print("Years in Credits only:", Credits_only)

print(f"{'Tax year':10}{'Taxable Sales and Purchases':>25}{'Credits':>20}")
for key, value in Credits.items() :
    if key in Sales.keys() :
        print(f"{key:20}{Sales[key]:>20,}{value:>20,}")
```
