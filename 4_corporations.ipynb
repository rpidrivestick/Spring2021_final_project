```python
import csv, operator, datetime

fn = "Active_Corporations___Beginning_1800.csv"
f = open(fn, 'r')
## uses csv Dictreader
reader = csv.reader(f)
CountYear = 2018
EndYear = 2021
filed_dict = {}

for y in range(CountYear, EndYear) :
    # to align with sales tax year, first 10 months of reporting year are months 3-12 in year y, last 2 months of reporting year 
    #   are months 1 and 2 in year 'y + 1'
    CountYear2 = int(y) + 1
    # for check, print the calendar years related to the reporting year
    print(y, CountYear2)
    n = 0
    # clear out the Periods dictionary each iteration - built with data from each year, and then loaded into filed_dict dictionary at end of loop
    Periods = {}

    # must reopen the file each loop or we will get only the first loop in our results
    f = open(fn, 'r')
    reader = csv.reader(f)

    for row in reader:
        filing_dt = row[2]
        if n > 0 and filing_dt != '':

            file_dt =  datetime.datetime.strptime(filing_dt, "%m/%d/%Y")
            filing_mo = file_dt.month
            filing_day = file_dt.day
            filing_year = file_dt.year

            if filing_year == y and filing_mo > 2 :
                # adjust the month by 2 to align with sales tax periods
                mo_p = filing_mo - 2
                # zfill so that periods in dictionary are sorted correctly
                mo = "Period " + str(mo_p).zfill(2)
                if mo in Periods.keys() :
                    Periods[mo] += 1
                else:
                    Periods[mo] = 1
            elif filing_year == CountYear2 and filing_mo <= 2 :
                # adjust January and February to periods 11 and 12 to align with sales tax periods
                mo_p = filing_mo + 10
                # zfill so that periods in dictionary are sorted correctly
                mo = "Period " + str(mo_p).zfill(2)
                if mo in Periods.keys() :
                    Periods[mo] += 1
                else:
                    Periods[mo] = 1    
        if n % 500000 == 0 :
            print(n)
        #if n > 500000 :
            #break
        n += 1
    
    # define TotalFiled / reset each loop through year range
    TotalFiled = 0

    # calculate grand total for the reporting year
    for key, value in Periods.items():
        n_filed = TotalFiled
        n_add = int(value)
        TotalFiled = n_filed + n_add
    # adds the total as a key and value in the Periods dictionary - CountYear2 is used because this is the reporting year/sales tax year
    Periods['Year ' + str(CountYear2) + ' Total'] = f"{TotalFiled:,}"
    # test print
    # print('Year ', CountYear2, ' Total: ', f"{TotalFiled:,}")
    
    # test print
    # print(Periods)

    # sort the dictionary by the key
    sorted_Periods = sorted(Periods.items(), key = operator.itemgetter(0), reverse = False)
    # test print
    # print(sorted_Periods)

    # define a dictionary in the filed_dict dictionary as 'dict ' + reporting year
    filed_dict['dict_' + str(CountYear2)] = sorted_Periods

# prints the complete filed_dict dictionary
print(filed_dict)
```
